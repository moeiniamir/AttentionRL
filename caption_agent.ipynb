```python
config = GPT2Config(vocab_size=10_000, n_positions=128, n_embd=256, n_layer=8, n_head=4, eos_token_id=2, bos_token_id=0, add_cross_attention=True)
policy_nework = GPT2LMHeadModel(config)
value_nework = GPT2LMHeadModel(config)
```

```python
with open('all_text.txt', 'w') as f:
  for k, v in captions.items():
    for sent in v:
      f.write(sent + '\n')
```

```python
tokenizer = ByteLevelBPETokenizer()
```

```python
# Customize training
tokenizer.train(files='all_text.txt', vocab_size=10_000, min_frequency=2, special_tokens=[
    "<s>",
    "<pad>",
    "</s>",
    "<unk>",
    "<mask>",
])
```
